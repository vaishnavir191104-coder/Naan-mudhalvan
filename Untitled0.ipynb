```python
import gradio as gr

def generate_site(idea):

    html = f"""
<!DOCTYPE html>
<html>
<head>
<title>AI Generated Website</title>

<style>

body {{
font-family:Arial;
margin:0;
background:#f4f4f4;
}}

header {{
background:#222;
color:white;
padding:15px;
display:flex;
justify-content:space-between;
}}

header a {{
color:white;
margin:10px;
text-decoration:none;
}}

.hero {{
background:linear-gradient(45deg,#4CAF50,#2E7D32);
color:white;
padding:80px;
text-align:center;
}}

.container {{
padding:30px;
}}

.products {{
display:grid;
grid-template-columns:repeat(auto-fit,minmax(220px,1fr));
gap:20px;
}}

.card {{
background:white;
padding:20px;
border-radius:10px;
box-shadow:0 0 10px rgba(0,0,0,0.1);
}}

button {{
background:#4CAF50;
color:white;
border:none;
padding:10px;
border-radius:5px;
cursor:pointer;
}}

footer {{
background:#222;
color:white;
text-align:center;
padding:15px;
margin-top:30px;
}}

</style>

</head>

<body>

<header>
<h2>AI Website</h2>
<div>
<a href="#">Home</a>
<a href="#">Products</a>
<a href="#">Contact</a>
</div>
</header>

<div class="hero">
<h1>AI Generated Website</h1>
<p>{idea}</p>
</div>

<div class="container">

<h2>Our Products</h2>

<div class="products">

<div class="card">
<h3>Silk Saree</h3>
<p>Premium silk saree collection</p>
<button>Buy Now</button>
</div>

<div class="card">
<h3>Cotton Saree</h3>
<p>Comfortable cotton sarees</p>
<button>Buy Now</button>
</div>

<div class="card">
<h3>Designer Saree</h3>
<p>Latest designer collection</p>
<button>Buy Now</button>
</div>

</div>

<h2>Contact Us</h2>
<p>Email: shop@gmail.com</p>
<p>Phone: 9876543210</p>

</div>

<footer>
<p>Generated using AI HTML Quick Styler</p>
</footer>

</body>
</html>
"""

    return html, html


demo = gr.Interface(
    fn=generate_site,
    inputs=gr.Textbox(label="Enter your website idea"),
    outputs=[
        gr.HTML(label="Live Website Preview"),
        gr.Code(label="Generated HTML Code")
    ],
    title="Advanced AI HTML Website Generator",
    description="Enter your idea and generate a complete website with preview."
)

demo.launch()
```
